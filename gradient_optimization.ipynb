```python
%pylab inline
from sympy import *
init_printing()
# from matplotlib import pyplot as plt
```

```python
w1, w2 = symbols('w1 w2')
func = exp(-5*(w1-1)**2-10*(w2-1)**2)
```

```python
func
```

```python
gradiant = [diff(func, w1),diff(func, w2)]
```

```python
gradiant
```

```python
plotting.plot3d(func, (w1,0,2), (w2,0, 2))
```

```python
p1 = plot_implicit(Eq(func, 0), show=False)
cstep=0.075
[p1.extend(plot_implicit(Eq(func, x), show=False)) for x in np.arange(cstep, 1, cstep)]
p1.xlim = p1.ylim = (0,2)
#p1.show()

p1.show()
fig = p1._backend.fig
ax = p1._backend.ax




step_size = 0.02
steps = 15
start_point = np.array([0.6, 1.3])
# colors = ['r' for x in range(steps)]
def colors(x):
    colorvs = ['r', 'g']
    return colorvs[x%len(colorvs)]
fvals = []
for x in range(steps):
    subs = dict(zip(func.free_symbols, start_point))
    grad_val_at_point = np.array([N(g, subs=subs) for g in gradiant])
    func_val_at_point = N(func, subs=subs)
    fvals += [float(func_val_at_point)]
    print start_point, func_val_at_point
    difference = step_size*grad_val_at_point
    next_point = start_point +  difference
    #fig2 = plt.figure()
    ax.arrow(float(start_point[0]), float(start_point[1]) , float(difference[0]), float(difference[1]), 
              head_width=0.03, head_length=0.02, fc=colors(x), ec=colors(x))
    start_point = next_point
fig
```

```python
plt.plot(fvals)
```
