### Data Generation

```python
data_size = 50000
multp = -1
noisemultp = 6
x1 = np.random.random_integers(-10, 10, data_size) + noisemultp*np.random.randn(data_size)
x2 = multp*x1 + noisemultp*np.random.randn(data_size)
#x2 = x1 #np.random.random_integers(-10, 10, data_size) + noisemultp*np.random.randn(data_size)


#uncomment the following for correlated data
#correlated_data = np.random.multivariate_normal([0.0, 0.0], np.matrix('1, -1;-1, 1'), data_size)
#x1 = correlated_data[:, 0]
#x2 = correlated_data[:, 1]

#uncomment the following for uncorrelated data
#uncorrelated_data = np.random.multivariate_normal([0.0, 0.0], np.matrix('10, 0;0, 10'), data_size)
#x1 = uncorrelated_data[:, 0]
#x2 = uncorrelated_data[:, 1]
```

### Plot data

```python
#prepare figure and axis
fig = plt.figure(1)
ax = SubplotZero(fig, 111)
fig.add_subplot(ax)
for direction in ["xzero", "yzero"]:
    ax.axis[direction].set_axisline_style("-|>")
    ax.axis[direction].set_visible(True)
for direction in ["left", "right", "bottom", "top"]:
    ax.axis[direction].set_visible(False)

#plot our data
ax.scatter(x1, x2, marker='.')
##plot the red line
#ax.plot(ax.get_xlim(),[x*multp for x in ax.get_xlim()], 'r')
ax.text(ax.get_xlim()[1] + 2,-.5,r'$x_1')
ax.text(-.5,ax.get_ylim()[1] + 2,r'$x_2')

```

## Now let's do some PCA

### Subtract mean

```python
x1 = x1 - np.mean(x1)
x2 = x2 - np.mean(x2)
```

### Calculate covariance matrix:

```python
cov_matrix = np.array([[np.mean([a*b for a,b in zip(d1, d2)]) for d1 in [x1, x2]] for d2 in [x1, x2]])
```

```python
#cov_matrix[0,1] = cov_matrix[1,0] = 0

print cov_matrix
```

### Eigen-value decomposition

#### Eigen-vectors are normalized.

```python
ws, us = eig(cov_matrix)
##sort eigen values and eigen vectors
#idx = ws.argsort()[::-1]
#ws = ws[idx]
#us = us[idx]
```

```python
projected_data = [us * np.matrix([c1, c2]).T for c1, c2 in zip(x1, x2)]
s1, s2 = [np.matrix([float(p[dim, :]) for p in projected_data]) for dim in [0, 1]]
```

```python
#Calculate the new axes
x_axis_before = np.matrix('-40, 40; 0, 0')
y_axis_before = np.matrix('0, 0; -60, 60')
print x_axis_before
print y_axis_before
```

```python
x_axis_projected = us * x_axis_before
y_axis_projected = us * y_axis_before
```

```python
def make_cartesian(ax):
    for direction in ["xzero", "yzero"]:
        ax.axis[direction].set_axisline_style("-|>")
        ax.axis[direction].set_visible(True)
    for direction in ["left", "right", "bottom", "top"]:
        ax.axis[direction].set_visible(False)

def create_line(axn, mat, col, label):    
    axn.arrow(mat[0, 0], mat[1, 0], mat[0, 1]-mat[0, 0], mat[1, 1] - mat[1, 0],
              label=label, fc=col, ec=col, head_width=3, head_length=4)
    return Line2D(mat[0, :], mat[1, :], color=col, lw=3, label=label)

fign = plt.figure(1)
axn = SubplotZero(fign, 211)
fign.add_subplot(axn)
make_cartesian(axn)

xax0 = create_line(axn, x_axis_before, 'b', r'$x1$')
yax0 = create_line(axn, y_axis_before, 'c', r'$x2$')
xax0 and axn.add_line(xax0)
yax0 and axn.add_line(yax0)

xax_proj = create_line(axn, x_axis_projected, 'r', r'$s1$')
yax_proj = create_line(axn, y_axis_projected, 'y', r'$s2$')
xax_proj and axn.add_line(xax_proj)
yax_proj and axn.add_line(yax_proj)

axn.scatter(x1, x2, marker='^', c='c', label='Original data')
axn.legend()

axn = SubplotZero(fign, 212)
fign.add_subplot(axn)
make_cartesian(axn)


xax0 = create_line(axn, x_axis_before, 'b', r'$x1$')
yax0 = create_line(axn, y_axis_before, 'c', r'$x2$')
xax0 and axn.add_line(xax0)
yax0 and axn.add_line(yax0)

xax_proj = create_line(axn, x_axis_projected, 'r', r'$s1$')
yax_proj = create_line(axn, y_axis_projected, 'y', r'$s2$')
xax_proj and axn.add_line(xax_proj)
yax_proj and axn.add_line(yax_proj)
axn.scatter(s1, s2, marker='v', c='y', label='Projected data')
axn.legend()
#axn.plot(axn.get_xlim(),[multp*x for x in axn.get_xlim()], 'g')
```
